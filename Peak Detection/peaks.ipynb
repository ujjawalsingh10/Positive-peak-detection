```python
import pandas as pd
```

```python
df = pd.read_csv('data_file.csv')
```

```python
df.isnull().values.sum()
```

```python
import matplotlib.pyplot as plt
plt.plot(df)
plt.show()
```

```python
x = df.iloc[:,0]
x.to_numpy()
```

```python
import numpy as np
from scipy.signal import find_peaks
peaks, _ = find_peaks(x, height = 0)
plt.plot(x)
plt.plot(peaks, x[peaks], "o")
plt.show()
```

```python
num_positive_peaks = len(peaks)
```

```python
print(f"Number of positive peaks in the dataset:", num_positive_peaks)
```
